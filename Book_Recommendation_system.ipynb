```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=";", error_bad_lines=False, encoding='latin-1')
```

```python
ratings
```

```python
ratings.shape
```

```python
ratings.rename(columns={
    "User-ID":  "user_id",
    "book_rating": "rating"
}, inplace = True)
```

```python
ratings
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
y= x[x].index
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books
```

```python
number_rating = ratings_with_books.groupby('title')['Book-Rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating
```

```python
final_rating = final_rating[final_rating['Book-Rating_y'] >= 50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'], inplace = True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values= 'Book-Rating_x')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[111,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
distance
```

```python
def recommand_book(book_name):
  book_index = np.where(book_pivot.index == book_name)[0][0]
  distance, suggestion = model.kneighbors(book_pivot.iloc[book_index,:].values.reshape(1,-1), n_neighbors=6 )

  for i in range(len(suggestion)):
    books = book_pivot.index[suggestion[i]]
    for j in books:
      if j == book_name:
        print(f"You searched '{book_name}'\n")
        print("The suggestion books are: \n")
      else:
        print(j)
```

```python
recommand_book("Exclusive")
```
